## DataFrame

DataFrame is used to create tables in python.

*   in dataframe there is columns
*   and rows
*   DataFrame() function is used in dataframe
*data frame is 2D array






### Different ways to create a dataframe

*  Through reading csv file
*Through reading excel file
*Through python Dictionary
*Through python tuple list
*Through python dictionary list

   




```python
# Dictionary
import pandas as pd

data = {
    'id':[1,2,3,4,5,6],
    'name':["sona","angel","kavya","aiswarya","ansif","sarang"],
    'age':[23,23,25,23,21,26],
}
df = pd.DataFrame(data)
df
```

```python
# tuple list
import pandas as pd

data = [
    (1,"sona",23),
    (2,"angel",23),
    (3,"kavya",25),
    (4,"ansif",21)
]
df = pd.DataFrame(data,columns=["id","Name","age"])
df
```

```python
#dictionary list
data = [
    {'id':1,'name':"sona",'age':23},
     {'id':2,'name':"angel",'age':23},
     {'id':3,'name':"ansif",'age':23},

]
df = pd.DataFrame(data)
df
```
