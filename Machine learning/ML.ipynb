## simple linear regression


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

area=np.array([1500,2000,2500,3000,3500]).reshape(-1,1)
price=np.array([300000,400000,500000,600000,700000])

model=LinearRegression()
model.fit(area,price)

predicted_price=model.predict([[3000]])
print(predicted_price)

print(f'slop(coefficient):{model.coef_[0]}',predicted_price)
print(f'intercept:{model.intercept_}')

predicted_price=model.predict(area)
plt.scatter(area,price,color='green',label='actual prices')
plt.plot(area,predicted_price,color='blue',label='Regression Line')
plt.xlabel('area|(squre feet)')
plt.ylabel('price(dollers)')
plt.title('simple linear regression:area v/s price')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

engine_size = np.array([1000, 1500, 2000, 2500, 3000]).reshape(-1, 1)
car_price = np.array([20000, 25000, 30000, 35000, 40000])

model=LinearRegression()
model.fit(engine_size,car_price)

pred=model.predict([[2500]])
print(pred)

print(f'slop(coefficient):{model.coef_[0]}',pred)
print(f'intercept:{model.intercept_}')

pred=model.predict(engine_size)
plt.scatter(engine_size,car_price,color='green',label='actual prices')
plt.plot(engine_size,pred,color='blue',label='Regression Line')
plt.xlabel('engine_size (in cubic centimeters)')
plt.ylabel('car_price(dollers)')
plt.title('simple linear regression: engine_size v/s car_price')
plt.legend()
plt.show()


```

```python
model.score(engine_size,car_price)
```

## logistic regression

*logistic regression is part of classification in supervised learning

*it is s shaped

*then why its called logistic regression because it has same process as linear regression the only difference is in last part,in logistic regression use sigmoid function(to convert into logistic regression probability)


```python
#example 1
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

#generating a synthetic dataset
np.random.seed(0) # .random function is used for generating randomly
                  #and .seed funcuion is used to to get same values or constantly what we generated randomly
x=np.concatenate([np.random.normal(2,1,50),np.random.normal(6,1,50)]).reshape(-1,1)
y=np.array([0]*50+[1]*50)

#train a logistic regression model
model=LogisticRegression()
model.fit(x,y)
```

```python
#plotting the data points
plt.figure(figsize=(10,6))
plt.scatter(x,y,color='red',label='data points')

#plot the logistic regression curve
x_test=np.linspace(x.min(),x.max(),1000).reshape(-1,1)
y_prob=model.predict_proba(x_test)[:,1]
plt.plot(x_test,y_prob,color='blue',linewidth=2,label='logistic regression curve')

#add label and legend
plt.xlabel('feature')
plt.ylabel('probability')
plt.title('Logistic regression')
plt.legend()

#show the plot
plt.show()
```

```python
#example 2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression,LinearRegression

#generating a synthetic dataset
np.random.seed(0)
x=np.concatenate([np.random.normal(2,1,50),np.random.normal(6,1,50)]).reshape(-1,1)
y=np.array([0]*50+[1]*50)

#train a logistic regression model
logistic_model=LogisticRegression()
logistic_model.fit(x,y)
```

```python
#train a Linear regression model
linear_model=LinearRegression()
linear_model.fit(x,y)
```

```python
#plotting the data points
plt.figure(figsize=(10,6))
plt.scatter(x,y,color='red',label='data points')

#plot the logistic regression curve
x_test=np.linspace(x.min(),x.max(),1000).reshape(-1,1)
y_logistic_prob=model.predict_proba(x_test)[:,1]
plt.plot(x_test,y_logistic_prob,color='blue',linewidth=2,label='logistic regression curve')

#plot the linear regression line
y_linear_pred=linear_model.predict(x_test)
plt.plot(x_test,y_linear_pred,color='green',linestyle='--',linewidth=2,label='linear regression line')

#add label and legend
plt.xlabel('feature')
plt.ylabel('output')
plt.title('Logistic vs Linear regression')
plt.legend()

#show the plot
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Load the iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a logistic regression model
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Compute the confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Display the confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=iris.target_names)
disp.plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix - Iris Dataset")
plt.show()
```
