```python
import pandas as pd
covi=pd.read_csv('/content/Covid_19 Kerala.csv2020.csv')
covi
```

```python
covi.info()
```

```python
del covi['Date']
```

```python
del covi['Time']
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
labelencoder
```

```python
covi['State']=labelencoder.fit_transform(covi['State'])
covi
```

```python
covi['Conformed Indian Nationals']=labelencoder.fit_transform(covi['Conformed Indian Nationals'])
covi
```

```python
covi['Conformed Foreign Nationals']=labelencoder.fit_transform(covi['Conformed Foreign Nationals'])
covi
```

```python
covi.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
corr_covi = covi.corr()
plt.figure(figsize=(5, 5))
sns.heatmap(corr_covi, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
y=covi[['Conformed']]
x=covi[['Conformed Indian Nationals','Conformed Foreign Nationals','Cured','Deaths',]]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.3)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(xtrain,ytrain)
```

```python
model.predict([[14,2,3,0]])
```

```python
y_predicted=model.predict(xtest)
y_predicted
```

```python
model.score(xtest,ytest)
```
