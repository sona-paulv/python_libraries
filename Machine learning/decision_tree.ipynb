# salary

```python
import pandas as pd
data=pd.read_csv('/content/salaries (2).csv')
data
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['company']=le.fit_transform(data['company'])
data['job']=le.fit_transform(data['job'])
data['degree']=le.fit_transform(data['degree'])
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
y=data['salary_more_then_100k']
x=data.drop('salary_more_then_100k',axis=1)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
 from sklearn.tree import DecisionTreeClassifier
 model=DecisionTreeClassifier()
 model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred = y_pred.reshape(-1, 1)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

## titanic

```python
#decision tree classifier for classification
import pandas as pd
data=pd.read_csv('/content/Titanic (1) (1).csv')
data
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
labelencoder
```

```python
data['Sex']=labelencoder.fit_transform(data['Sex'])
data['Embarked']=labelencoder.fit_transform(data['Embarked'])

data
```

```python
data['Age'].fillna(data['Age'].mean(),inplace=True)
data
```

```python
del data['Cabin']
```

```python
data['Embarked'].fillna(data['Embarked'].median(),inplace=True)
data
```

```python
del data['Ticket']
data
```

```python
del data['Name']
data
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
corr_data = data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_data, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
y=data[['Survived']]
x=data[['Pclass','Sex','Parch','Fare']]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
model.predict([[1,0,0,71.28330]])
```

```python
model.score(x_test,y_test)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

## petrol consumption

```python
import pandas as pd
data=pd.read_csv('/content/petrol_consumption.csv')
data.head()
```

```python
data.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
corr_data = data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_data, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
del data['Paved_Highways']
```

```python
y=data[['Petrol_Consumption']]
x=data[['Petrol_tax','Average_income','Population_Driver_licence(%)']]
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(xtrain,ytrain)
```

```python
model.predict([[9.00,4092,0.572]])
```

```python
ypred=model.predict(xtest)
ypred
```

```python
from sklearn.metrics import r2_score,mean_squared_error
r2socre=r2_score(ytest,ypred)
r2socre
```

```python
meansqure=mean_squared_error(ytest,ypred)
meansqure
```

```python
data.boxplot()
```

```python
def remove_outliers_iqr(df, exclude_column):
    for col in df.columns:
        # Skip the specified column
        if col == exclude_column:
            continue
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # Filter the DataFrame to remove rows with outliers in each column
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df

# Specify the column to exclude
df_cleaned = remove_outliers_iqr(data, exclude_column='Petrol_Consumption')
df=df_cleaned
```

```python
df.head()
```

```python
df.boxplot()
```

```python
y=df[['Petrol_Consumption']]
x=df[['Petrol_tax','Average_income','Population_Driver_licence(%)']]
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(xtrain,ytrain)
```

```python
model.predict([[9.00,4092,0.572]])
```

```python
model.score(xtest,ytest)
```

```python
ypred=model.predict(xtest)
ypred
```

```python
from sklearn.metrics import r2_score,mean_squared_error
r2socre=r2_score(ytest,ypred)
r2socre
```

```python
meansqure=mean_squared_error(ytest,ypred)
meansqure
```
